# Parte 2

```python
df_concat = df_titulos.merge(df_originals, how="inner", right_on="Title", left_on="title")
df_concat.head()
```

## Justificación del merge:

- He decido realizar un inner en el merge porque considero que es la forma de obtener menos nulos. Además tomo de referencia el dataframe de netflix_titles porque es el que más datos tiene. Los he unido a través de la columna title ya que he verificado que es unica y común entre los dos dataframes.
