# Conclusiones generales

En este dataframe hay titulos de peliculas y series de Netflix.

Existen 13 columnas y 8807 peliculas y series. La mayoria son de tipo objeto. Existen 11 columnas de tipo objeto y 2 int.

Hay valores nulos pero no duplicados. Apuntar columnas de nulos

Exiten 6131 peliculas y 2676 series.

En cuanto a las clasificaciones de edad la mayoria de peliculas estan clasificadas con TV-MA. Son 2062 filas que equivale a un 23.42% del total. La clasficación de edad de las series que más porcentaje tiene es TV-MA con un total de 1145 que equivale a un 13.01% del total. Dentro de esta categoría existen datos, en concreto 3 filas, que nos generarán inconsistencias al hacer un analisís en el futuro, estos datos tienen que ver con la duración de las peliculas.

El director que más peliculas o tv shows es Rajiv Chilaka con 19 pelliculas.

En un futuro se podría poner como indice la columna 'show_id'. Resulta ser el identificador de cada serie y son valores unicos.

El pais con más producciones es United States con 2818 que equivale a un 35.33% del total.
